```python
def running_mean(x, N):
    cumsum = numpy.cumsum(numpy.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)

    
    
```

```python
df = pd.read_csv('data_temp/seoul_data.csv')
print(df['avg_temp'].shape)

# moving average over the temprature values

df['MA'] = df['avg_temp'].rolling(window=5).mean()

# Plotting city temprature data after with moving average values
plt.plot(df['year'], df['MA'])
plt.title("Seoul city temprature by years")
plt.xlabel('Years')
plt.ylabel('Temprature')
plt.show()
```

```python
df_global = pd.read_csv('data_temp/global_data.csv')
print(df_global['year'])
df_global['MA'] = df_global['avg_temp'].rolling(window=5).mean()

plt.plot(df_global['year'], df_global['MA'])
plt.title("Seoul city temprature by years")

plt.xlabel('Years')
plt.ylabel('Temprature')
plt.show()
```

## Local and Global temprature

```python
n = len([i for i in df_global['year'] if i <= df['year'].min()])
print(n)
plt.title("Local and Global temprature trends")

plt.plot(df['year'], df['MA'], label = 'Local')
plt.plot(df_global['year'][n:], df_global['MA'][n:], label = 'Global')
plt.xlabel('Years')
plt.ylabel('Temprature')
plt.legend(loc = 'upper left')
plt.savefig('Local_and_Global_temp.png')
plt.show()
```
