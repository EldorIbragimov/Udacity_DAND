### Importing libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Reading data and plotting for Local temprature

```python
# Reading Local(Seoul city) data from a directory
df = pd.read_csv('data_temp/seoul_data.csv')

# moving average over the temprature values
df['MA'] = df['avg_temp'].rolling(window=5).mean()

# Plotting city temprature data with moving average values
plt.plot(df['year'], df['MA'])
plt.title("Seoul city temprature trends from 1839 to 2015")
plt.xlabel('Years')
plt.ylabel('Temprature')
plt.savefig('Seoul_city_temp.png')

plt.show()
```

### Reading data and plotting for Global temprature

```python
df_global = pd.read_csv('data_temp/global_data.csv')
#print(df_global['year'])
df_global['MA'] = df_global['avg_temp'].rolling(window=5).mean()

plt.plot(df_global['year'], df_global['MA'])
plt.title("Glocal temprature by years")

plt.xlabel('Years')
plt.ylabel('Temprature')
plt.savefig('Global_data_temp.png')
plt.show()
```

## Local and Global temprature

```python
n = len([i for i in df_global['year'] if i <= df['year'].min()])
#print(n)
plt.title("Local(Seoul City) and Global temprature trends")
plt.plot(df['year'], df['MA'], label = 'Local')
#plt.text(df['year'][1], df['MA'][8], 'min temp = '+str(df['MA'][4]), fontsize = 12)
#plt.plot(df['year'][4], df['MA'][4], marker = 'o', markersize = 6)

plt.plot(df_global['year'][n:], df_global['MA'][n:], label = 'Global')
plt.xlabel('Years')
plt.ylabel('Temprature')
plt.legend(loc = 'upper left')

plt.savefig('Local_and_Global_temp.png')

plt.show()
```
